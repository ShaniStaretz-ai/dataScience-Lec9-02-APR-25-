```python
df.set_index('Sport Type').groupby('Sport Type', group_keys=False).apply(lambda x: x.nlargest(2, 'Height(cm)')) .reset_index()
```

```python
groups=df.groupby(['Sport Type','Country'])
groups.max().loc['Soccer','USA']
groups.max().xs(level='Sport Type',key ='Soccer')
df[df['Sport Type'].isin(['Tennis','Soccer'])].groupby(['Sport Type','Country']).max()
groups.max().swaplevel().sort_values(by=['Sport Type','Country'])

height_wins_stat=groups.agg({
    'Height(cm)': ['min','max','mean','std'],
    'Wins':['min','max','mean','std']})
height_wins_stat[('Height(cm)', 'std')]=height_wins_stat[('Height(cm)', 'std')].fillna(0)
height_wins_stat[('Wins', 'std')]=height_wins_stat[('Wins', 'std')].fillna(0)
height_wins_stat
```
