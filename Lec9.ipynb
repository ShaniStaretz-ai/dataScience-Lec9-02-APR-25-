```python
import numpy as np
import pandas as pd
from IPython.display import Markdown,display
def printmd(message,data):
    text="**<u>"+message+"</u>**"
    display(Markdown(text),data)
df = pd.read_csv("mpg.csv")
groups1=df.groupby('model_year')['horsepower'].max()
groups2=df.groupby('model_year')['acceleration'].mean()
result=pd.concat([groups1,groups2],axis=1)
result
```
